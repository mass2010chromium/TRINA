```python
async def checkFeedback(id):
    print("hehe")
    while not a.server['UI_FEEDBACK'][str(id)]['REPLIED'].read():
        continue
    print("haha")
    return a.server['UI_FEEDBACK'][str(id)]['MSG'].read()
    
```

```python
async def addConfirmation(title,text):
    id = '$'+ uuid.uuid1().hex
    msg = {'funcName':'addConfirmation','args':{'id':str(id),'title':title,'text':text}}
    commandQueue = a.server["UI_END_COMMAND"].read()
    commandQueue.append(msg)
    a.server['UI_FEEDBACK'][str(id)] = {'REPLIED':False, 'MSG':''}
    a.server["UI_END_COMMAND"] = commandQueue
    print("commandQueue", commandQueue)
    time.sleep(0.0001) 
    val = await checkFeedback(str(id))
    print("I love you")
    return  val
```
